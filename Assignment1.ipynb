## Insert Data

```python
import requests

def get_dataset(url):
  # get the content under this url
  result = requests.get(url)

  # print the first 200 characters
  print(result.text[:200])

  # return the retrieved text
  return result.textb
```

```python
from bson.json_util import loads

def load_txt_to_bson(txt):
  data = []
  # split by new line (since each object is divided using a new line)
  for r in txt.split("\n"):
      # if
      if len(r) > 0:
        data.append(loads(r))
  return data
```

```python
dataset = {}
dataset['url'] = "https://raw.githubusercontent.com/neelabalan/mongodb-sample-dataset/refs/heads/main/sample_supplies/sales.json"
dataset['bson_dataset'] = load_txt_to_bson(get_dataset(dataset['url']))
```

```python
# Remove '_id' field from each document to prevent duplicates
for doc in dataset['bson_dataset']:
    doc.pop('_id', None)  # Safely remove '_id' if it exists

# Now insert the cleaned dataset into MongoDB
collection.insert_many(dataset['bson_dataset'])
```
